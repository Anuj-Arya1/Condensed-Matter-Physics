# Metal–Insulator Transition in Ladder Tight-Binding Model

## Objective
To plot the band structure for different values of the ratio \( t'/t \) and identify the transition from metallic to insulating behavior.

---

## Band Structure

The energy dispersion for the ladder model is given by:

\[
E_{\pm}(k) = -2t\cos(ka) \pm t'
\]

This results in two bands:
- Lower band: \( E_{-}(k) \)
- Upper band: \( E_{+}(k) \)

---

## Method

To study the transition, the band structure is plotted for a range of values of \( t'/t \), such as:

- \( t'/t = 0.5 \)
- \( t'/t = 1.0 \)
- \( t'/t = 1.5 \)
- \( t'/t = 2.0 \)
- \( t'/t = 2.5 \)

---

## Analysis

The nature of the system depends on whether the two bands overlap or are separated by a gap.

### Band edges:
- Maximum of lower band:  
  \[
  E_{-}^{\text{max}} = 2t - t'
  \]

- Minimum of upper band:  
  \[
  E_{+}^{\text{min}} = -2t + t'
  \]

---

## Metal–Insulator Condition

### Metallic Phase (Band Overlap)
\[
t' < 2t
\]

- Bands overlap
- No energy gap at the Fermi level
- System behaves as a metal

---

### Insulating Phase (Band Gap)
\[
t' > 2t
\]

- Bands are separated
- Finite energy gap exists
- System behaves as an insulator

---

## Transition Point

The transition occurs when the band gap closes:

\[
t' = 2t
\]

At this point:
- The upper and lower bands just touch
- The system changes from metallic to insulating behavior

---

## Conclusion

By plotting the band structure for different values of \( t'/t \), it is observed that the system undergoes a metal-to-insulator transition at \( t' = 2t \). This transition arises from the competition between band dispersion (controlled by \( t \)) and band splitting (controlled by \( t' \)).

```python
# Question 1
t = 1.0   # hopping along legs
t_prime_L = [0.5,1,1.5,2,2.5,3.0]   # different t'/t ratios
t_t_prime = [t/t_prime for t_prime in t_prime_L]
a = 1.0 

# k-space
k = np.linspace(-np.pi/a, np.pi/a, 1000)

for t_prime in t_prime_L:
    # Here E0 (on site term) is taken zero
    E_plus = -2*t*np.cos(k*a) + t_prime
    E_minus = -2*t*np.cos(k*a) - t_prime
    Gap =  - round(np.max(E_minus),3) + round(np.min(E_plus),3)
    if Gap > 0:
        phase = "Insulator"
    elif Gap < 0:
        phase = "Metal"
    else:
        phase = "Transition"
    print(f"t' = {t_prime:.2f}| t/t' = {t_t_prime[t_prime_L.index(t_prime)]:.2f} | Gap = {Gap:.3f} → {phase}")
    plt.plot(k, E_plus, label=r'$E_{+}$' + f"(t'={t_prime})")
    plt.plot(k, E_minus, linestyle='--', label= r'$E_{-}$' + f" (t'={t_prime})")

plt.xlabel("k")
plt.axvline(-np.pi/a, color='grey', linestyle='--', linewidth=1)
plt.axvline(np.pi/a, color='grey', linestyle='--', linewidth=1)
plt.ylabel("Energy")
plt.title("Band Structure of TB")
plt.legend()
plt.grid(False)
plt.show()
```

# Effect of Diagonal Hopping in Ladder Tight-Binding Model

## Objective
To study how the inclusion of diagonal hopping \( t'' \) affects the band structure and the metal–insulator transition in the ladder model.

---

## Modified Hamiltonian

In addition to:
- Nearest-neighbor hopping along legs: \( t \)
- Inter-leg (rung) hopping: \( t' \)

We now include:
- Diagonal hopping: \( t'' \)

The Hamiltonian in k-space becomes:

\[
H(k) =
\begin{pmatrix}
-2t\cos(ka) & -t' - 2t''\cos(ka) \\
-t' - 2t''\cos(ka) & -2t\cos(ka)
\end{pmatrix}
\]

---

## Band Structure

The energy bands are obtained by diagonalizing the Hamiltonian:

\[
E_{\pm}(k) = -2t\cos(ka) \pm \left( t' + 2t''\cos(ka) \right)
\]

---

## Method

- The band structure is plotted for different values of \( t'' \), keeping:
  \[
  |t''| < |t|,\; |t'|
  \]
- Typical values used:
  - \( t'' = 0 \)
  - \( t'' = 0.3 \)
  - \( t'' = -0.3 \)

---

## Band Edge Analysis

To determine whether the system is metallic or insulating, we compute:

- Maximum of lower band:
  \[
  E_-^{\text{max}} = \max(E_-(k))
  \]

- Minimum of upper band:
  \[
  E_+^{\text{min}} = \min(E_+(k))
  \]

- Band gap:
  \[
  \text{Gap} = E_+^{\text{min}} - E_-^{\text{max}}
  \]

---

## Metal–Insulator Condition

- **Metallic phase**:
  \[
  \text{Gap} < 0
  \]
  Bands overlap at different k-points.

- **Insulating phase**:
  \[
  \text{Gap} > 0
  \]
  Bands are separated by a finite gap.

- **Transition point**:
  \[
  \text{Gap} = 0
  \]

---

## Effect of Diagonal Hopping \( t'' \)

- The term \( 2t''\cos(ka) \) introduces **k-dependent inter-leg coupling**.
- This causes:
  - Distortion of band shapes
  - Asymmetry in dispersion
  - Shift in band extrema

---

## Physical Interpretation

- For \( t'' = 0 \), the system reduces to the standard ladder model.
- For \( t'' \neq 0 \):
  - The effective coupling between legs varies with k
  - The band gap is modified in a non-uniform way

- The sign of \( t'' \) plays a crucial role:
  - \( t'' > 0 \): enhances splitting at some k-points
  - \( t'' < 0 \): reduces splitting at some k-points

---

## Conclusion

The inclusion of diagonal hopping \( t'' \) significantly modifies the band structure by introducing k-dependent coupling. This leads to a shift in the metal–insulator transition point and alters the size of the band gap depending on the magnitude and sign of \( t'' \).

```python
# Question 2
t = 2
t_prime = 4
t_prime2 = [1e-6,0.6, 0.3, -0.3,-0.6]  
a = 1.0

def tbb_func(t_prime2,t_prime,t,a,par):
    for tpp in t_prime2:
        E_plus = []
        E_minus = []
        k = np.linspace(-np.pi/a, np.pi/a, 1000)
        for ki in k:
            # Hamiltonian with diagonal hopping
            if par == 'a':
                H = np.array([
                    [-2*t*np.cos(ki*a), -t_prime - tpp*np.exp(1j*ki*a)],
                    [-t_prime - tpp*np.exp(-1j*ki*a), -2*t*np.cos(ki*a)]
                ]) 
            elif par == 'b':
                H = np.array([
                    [-2*t*np.cos(ki*a), -t_prime - 2*tpp*np.cos(ki*a)],
                    [-t_prime - 2*tpp*np.cos(ki*a), -2*t*np.cos(ki*a)]
                ]) 
            else:
                return "Error"

            eigvals = np.linalg.eigvalsh(H)
            E_minus.append(eigvals[0])
            E_plus.append(eigvals[1])

        E_minus = np.array(E_minus)
        E_plus = np.array(E_plus)

        gap = np.min(E_plus)- np.max(E_minus)

        # Phase
        if gap > 0:
            phase = "Insulator"
        elif gap < 0:
            phase = "Metal"
        else:
            phase = "Transition"

        print(f"t'' = {tpp:.2f} | Gap = {gap:.3f} → {phase}")

        # Plot bands
        plt.plot(k, E_plus, label= r"$E_{+}$" + f" (t''={tpp})")
        plt.plot(k, E_minus, linestyle='--',label= r"$E_{-}$" + f"(t''={tpp})")

    plt.xlabel("k")
    plt.ylabel("Energy")
    plt.axvline(-np.pi/a, color='grey', linestyle='--', linewidth=1)
    plt.axvline(np.pi/a, color='grey', linestyle='--', linewidth=1)
    plt.title("Band Structure with Diagonal Hopping ")
    plt.legend()
    plt.grid(False)
    plt.show()

tbb_func(t_prime2,t_prime,t,a,'a')
tbb_func(t_prime2,t_prime,t,a,'b')
```

```python

# Question 3
t = 2.0  # in eV
t_prime_ = [-1.0, 1.0]

def contour_plot(t,t_p,part):
    # k-space grid (1st BZ)
    kx = np.linspace(-np.pi, np.pi, 200)
    ky = np.linspace(-np.pi, np.pi, 200)
    KX, KY = np.meshgrid(kx, ky)
    if part =='a':
        # Energy dispersion
        E = -2*t*(np.cos(KX) + np.cos(KY))
    elif part == 'b':
        E = -2*t*(np.cos(KX) + np.cos(KY)) - 4*t_p*np.cos(KX)*np.cos(KY)
    else:
        return 'ERROR!'

    # Flatten and sort energies
    E_flat = np.sort(E.flatten())
    
    # Electron filling values
    fillings = np.arange(0.5, 2.01, 0.25)

    plt.figure(figsize=(8,6))

    # Plot energy contour
    plt.contour(KX, KY, E, levels=20)

    # Fermi surface plotting
    for f in fillings:
        index = int(f/2 * (len(E_flat)-1))   # 2 states per unit cell
        Ef = E_flat[index]
        cs = plt.contour(KX, KY, E, levels=[Ef],linestyles='--', colors='r')
        plt.clabel(cs, fmt={Ef: f"f={f:.2f}"})
        print(f"Filling = {f:.2f} e/unit cell → Ef = {Ef:.3f}")
    
    plt.xlabel("$k_x$")
    plt.ylabel("$k_y$")
    plt.title("Square Lattice Band Structure & Fermi Surface (NN hopping)")
    plt.grid(False)
    plt.show()

# 3(a)
contour_plot(t,None,'a')

# 3(b)
for tp in t_prime_:
    contour_plot(t,tp,'b')
```

```python

def threeD_plot(t1, t2, a1, a2, E0, figsize=(10, 7), resolution=100):
    kx = np.linspace(-np.pi/a1, np.pi/a1, resolution)
    ky = np.linspace(-np.pi/a2, np.pi/a2, resolution)
    KX, KY = np.meshgrid(kx, ky)
    E = E0 - 2*t1*np.cos(KX*a1) - 2*t2*np.cos(KY*a2)

    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection='3d')

    surf = ax.plot_surface(KX, KY, E, cmap='plasma', linewidth=0,
                           antialiased=True, alpha=0.95)

    cbar = fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, pad=0.1)
    cbar.set_label('Energy (eV)', fontsize=12)

    ax.set_xlabel('kx (1/Å)', fontsize=11, labelpad=8)
    ax.set_ylabel('ky (1/Å)', fontsize=11, labelpad=8)
    ax.set_zlabel('E (eV)', fontsize=11, labelpad=8)
    ax.set_title(f'2D Tight-Binding Band Structure\n'
                 f't₁={t1}, t₂={t2}, a₁={a1}, a₂={a2}, E₀={E0}', fontsize=12)

    ax.view_init(elev=25, azim=-60)
    plt.tight_layout()
    plt.show()

threeD_plot(1,2,1,1,0)
```
